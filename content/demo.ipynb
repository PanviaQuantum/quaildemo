```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans-serif'
import numpy as np
```

```python
class quail:
    def __init__(self,filepath=None):
        self.debug = True
        self.programs = []
        self.figure = None
        self.axes = None
        self.load( filepath )

    def load(self, filepath ):
        if filepath == None:
            return
        commands = json.load( open( filepath ))
        circuit = program( commands )
        self.programs.append( circuit )
        self.figure = plt.figure()
        self.axes = self.figure.add_axes((0,0,2.5,6))
        self.axes.set_axis_off()

    def display(self):
        for circuit in self.programs:
            circuit.diagram(self.axes)
        plt.show()

```
